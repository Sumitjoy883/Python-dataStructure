# Q2. Explain the difference between mutable and immutable data types with examples?












```python
## 1. Mutable Data Types
These are data types whose values can be changed (modified) after creation without changing their identity (memory address).


## 2. Immutable Data Types
These are data types whose values cannot be changed after creation. Any modification creates a new object in memory.



## Mutable Data Types Example
my_list = [1, 2, 3]
print(id(my_list))  # Memory address before modification

my_list.append(4)  # Modifying the list
print(my_list)  # Output: [1, 2, 3, 4]
print(id(my_list))  # Memory address remains the same


## immutable Data Types Example
x = 10
print(id(x))  # Memory address before modification

x = x + 5  # Modifying the value
print(x)  # Output: 15
print(id(x))  # Memory address changes
```

# Q3. What are the main differences between lists and tuples in Python?

##. 1. Mutability
- List ‚Üí Mutable (can be modified after creation)
- Tuple ‚Üí Immutable (cannot be modified after creation)

## 2. Syntax
- List ‚Üí Created using square brackets [ ]
- Tuple ‚Üí Created using parentheses ( )

## 3. Performance
- Lists ‚Üí are slower because they allow modifications.
- Tuples ‚Üí are faster because they are immutable, making them more memory-efficient.

## 4. Memory Usage
- Lists ‚Üí consume more memory since they need space for dynamic resizing.
- Tuples ‚Üí consume less memory as they are fixed-size and require less overhead.



# Q4. Describe how dictionaries store data?
## How Dictionaries Store Data (Simple Explanation)
  A dictionary in Python is like a locker system where:

-Each key is a unique locker number.
- Each value is the item stored inside that locker.
# How it Works Internally?
1. Keys are converted into numbers (hash values) using a function.
2. Python finds a specific location (index) in memory to store the key-value pair.
3. When you search for a key, Python quickly looks up the memory location and retrieves the value.

# Q6.What is a string in Python, and how is it different from a list?

A string in Python is a sequence of characters enclosed in quotes (either single ' or double "). It is used to represent text in Python.

1. Data Types Inside
- String ‚Üí Can only store characters (text).

2. Mutability
- String ‚Üí Immutable, meaning you cannot change a character in a string once it‚Äôs created.

- List ‚Üí Mutable, meaning you can change, add, or remove elements.

3. Operations
- String ‚Üí You can concatenate strings or repeat them, but you can‚Äôt modify individual characters.

- List ‚Üí You can perform many operations like adding/removing elements and changing individual items.

4. Usage
-  String ‚Üí Primarily used to represent text.
- List ‚Üí Used to store a collection of items that can be of different types.

#Q10.What advantages do dictionaries offer over lists for certain tasks?
Dictionaries (dict) and lists (list) are both used for storing data, but dictionaries have some key advantages for certain tasks.

1. Faster Lookups (O(1) vs. O(n))
- Dictionaries use hash tables, so they provide fast key-based access (O(1) time).

- Lists require a linear search (O(n) time), which is slower for large datasets.

2. Key-Value Pair Storage
- Dictionaries store data in a structured way using keys and values.
- Lists store items in an ordered way, making access index-based
